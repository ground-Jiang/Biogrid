biogrid tab 3.0 foramt transformed into gene names(interactors) in txt

```python
from google.colab import files
import pandas as pd
import io

# 上传数据文件
uploaded = files.upload()

# 读取数据（替换为你的文件名）
file_name = next(iter(uploaded))
df = pd.read_csv(io.BytesIO(uploaded[file_name]), sep='\t')

# 提取两列基因并合并
interactors_a = df['Official Symbol Interactor A'].dropna().tolist()
interactors_b = df['Official Symbol Interactor B'].dropna().tolist()
all_interactors = interactors_a + interactors_b

# 去重并过滤SUPT16H
unique_genes = sorted({gene for gene in all_interactors if gene != 'SUPT16H'})

# 显示结果
print("Unique genes (excluding SUPT16H):")
for gene in unique_genes:
    print(gene)

# 可选：保存结果到文件
with open('filtered_genes.txt', 'w') as f:
    f.write('\n'.join(unique_genes))
files.download('filtered_genes.txt')
```

colab check whether there are overlapping gene names between two txt files

```python
# Step 1: Upload the files to Colab
from google.colab import files

# Upload the first file
uploaded_file1 = files.upload()
file1_name = list(uploaded_file1.keys())[0]  # Get the name of the first file

# Upload the second file
uploaded_file2 = files.upload()
file2_name = list(uploaded_file2.keys())[0]  # Get the name of the second file

# Step 2: Read gene names from both files
def read_gene_names(file_path):
    with open(file_path, 'r') as file:
        gene_names = set(line.strip() for line in file if line.strip())  # Use a set for faster lookups
    return gene_names

# Read gene names from both files
gene_names_file1 = read_gene_names(file1_name)
gene_names_file2 = read_gene_names(file2_name)

# Step 3: Find overlapping gene names
overlapping_genes = gene_names_file1.intersection(gene_names_file2)

# Step 4: Display the results
if overlapping_genes:
    print(f"There are {len(overlapping_genes)} overlapping gene names:")
    for gene in overlapping_genes:
        print(gene)
else:
    print("There are no overlapping gene names between the two files.")
```
